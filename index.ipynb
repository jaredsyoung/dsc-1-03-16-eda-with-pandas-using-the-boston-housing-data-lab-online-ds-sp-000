```python
df.nox.hist()
plt.title('dist of nox')
plt.xlabel('nox levels')
plt.ylabel('neighborhoods')
```

```python
#interested in examining how NOX values vary by socioeconomic measures of neighborhood

#1. black * lstat - look at data for high-proportion black neighborhoods and/or high-proportion low status

df1 = df.loc[(df["black"] >= 396.24) | (df["lstat"] >= 16.42)]
df1['nox'].describe()

# nox: mean = 0.591, median = 0.584, stddev = 0.115, iqr 0.493 to 0.693
```

```python
#2. average rooms per home in upper quartile AND owner occupied built prior to 1940 in upper quartile
df2 = df.loc[(df["rm"] >= 6.595) & (df["age"] >= 93.8)]
df2['nox'].describe()

# nox: mean = 0.584, median = 0.584, stddev = 0.128, iqr 0.472 to 0.679
```

```python
#3. tax in upper quartile OR home value in upper quartile
df3 = df.loc[(df["tax"] >= 666) | (df["medv"] >= 25)]
df3['nox'].describe()

# nox: mean = 0.583, median = 0.601, stddev = 0.115, iqr 0.481 to 0.693
```

```python
#Next, use histograms and scatterplots to see whether you observe differences 
#for the subsets of the data. Make sure to use subplots so it is 
#easy to compare the relationships.

import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Polygon

#1. top vs. bottom quartile black concentration
df4 = df['nox'].loc[(df["black"] >= 396.24)]
df5 = df['nox'].loc[(df["black"] <= 376.73)]

#2. top vs. bottom quartile lower status concentration
df6 = df['nox'].loc[(df["lstat"] >= 16.42)]
df7 = df['nox'].loc[(df["lstat"] <= 7.18)]

fig, axs = plt.subplots(2, 2)

# top quartile black
axs[0, 0].boxplot(df4)
axs[0, 0].set_title('nox - q3 - black')

# bottom quartile black
axs[0, 1].boxplot(df5, 1)
axs[0, 1].set_title('nox - q1 - black')

# top quartile lower status
axs[1, 0].boxplot(df6, 1)
axs[1, 0].set_title('nox - q3 - low status')

# bottom quartile lower status
axs[1, 1].boxplot(df7, 1)
axs[1, 1].set_title('nox - q1 - low status')


fig.subplots_adjust(left=0.08, right=0.98, bottom=0.05, top=0.9,
                    hspace=0.4, wspace=0.3)

plt.show()
```

In this project I examined the relationship between NOX levels and certain markers of socioecononic status, 
hypothesizing that lower status / lower income (as measured by home value) / racial minority neighborhoods
would have greater exposure to higher levels of NOX concentrations.  I looked at neighborhoods with top quartile black concentration or top quartile lower status concentration, as well as top quartile rooms per household for top quartile owner occupied homes from prior to 1940 (as a marker of inter-generational assets not captured by tax or home value), as well as top quartile tax rate or top quartile median home value and looked at measures of central tendancy and dispersion for each of these data cuts.  I found NOX for top quartile black or top quartile lower status to be much higher than for the overall dataset.  Further drill down on NOX level by top vs. bottom quartile comparison for black concentration showed that NOX levels are actually somewhat lower in neighborhoods with high black concentration.  However, NOX levels did appear much higher in neighborhoods with higher concentration of lower status.



# Variable Descriptions

This data frame contains the following columns:

#### crim  
per capita crime rate by town.

#### zn  
proportion of residential land zoned for lots over 25,000 sq.ft.

#### indus  
proportion of non-retail business acres per town.

#### chas  
Charles River dummy variable (= 1 if tract bounds river; 0 otherwise).

#### nox  
nitrogen oxides concentration (parts per 10 million).

#### rm  
average number of rooms per dwelling.

#### age  
proportion of owner-occupied units built prior to 1940.

#### dis  
weighted mean of distances to five Boston employment centres.

#### rad  
index of accessibility to radial highways.

#### tax  
full-value property-tax rate per $10,000.

#### ptratio  
pupil-teacher ratio by town.

#### black  
1000(Bk - 0.63)^2 where Bk is the proportion of blacks by town.

#### lstat  
lower status of the population (percent).

#### medv  
median value of owner-occupied homes in $10000s.
